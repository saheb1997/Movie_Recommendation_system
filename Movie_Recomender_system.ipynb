```python
import ast
```

```python
def convert(obj):
   L=[]
   for i in ast.literal_eval(obj) :
       L.append(i['name'])
   return L 
```

```python
def convert2(obj):
   L=[]
   for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
           L.append(i['name'])
           break
   return L 
```

```python
def convert3(obj):
   L=[]
   counter=0
   for i in ast.literal_eval(obj) :
       if(counter!=3):
           L.append(i['name'])
           counter+=1
       else:
           break
   return L 
```

```python
def convert4(string):
        return string.split()
```

```python
def remove_space(s):
     l=[]
     for i in s:
        l.append(i.replace(" ", ""))
     return l;
```

```python
#here we take only dictonary value of name
movies['genres']=movies['genres'].apply(convert)

#for the cast we are the first three character 
movies['cast']=movies['cast'].apply(convert3)

#for keywords also we do the same
movies['keywords']=movies['keywords'].apply(convert)

movies['crew']=movies['crew'].apply(convert2)

movies['overview']=movies['overview'].apply(convert4)
```

```python
movies['genres']=movies['genres'].apply(remove_space)
movies['cast']=movies['cast'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
movies['crew']=movies['crew'].apply(remove_space)
```

```python
movies['Tag']=movies['genres']+movies['overview']+movies['cast']+movies['keywords']+movies['crew']
```

```python
new_df['Tag']=new_df['Tag'].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(  ps.stem(i))
    return " ".join(y)
```

```python
new_df['Tag']=new_df['Tag'].apply(stem)
```

```python
new_df['Tag']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['Tag']).toarray() 
```

```python
vector.shape
```

```python
for i in cv.get_feature_names_out():
    print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
for i in similarity[0]:
    print(i)
```

```python
val=list(enumerate(similarity[0]))
```

```python
for i in sorted(val,reverse=True,key=lambda x:x[1])[1:6]:
    print(i[0])
```

```python
def recomended(movies):
    movie_index=new_df[new_df['title']==movies].index[0]
    print(type(movie_index))
    distance=similarity[movie_index]
    
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```
